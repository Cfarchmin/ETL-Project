```python
#list all states
states=['AL','AK','AZ','AR','CA','CO','CT','DE','FL','GA',
'HI','ID','IL','IN','IA','KS','KY','LA','ME','MD',
'MA','MI','MN','MS','MO','MT','NE','NV','NH','NJ',
'NM','NY','NC','ND','OH','OK','OR','PA','RI','SC',
'SD','TN','TX','UT','VT','VA','WA','WV','WI','WY']

#states=['AL','AK']
```

```python
#create sections of api URL
beginning = "https://api.eia.gov/series/?api_key=4cdcec72db55cf889f2167eaf6e562ae&series_id=ELEC.GEN.ALL-"
end = "-99.A"
```

```python
#define columns
Columns = ["Period","Value"]
```

```python
#loop through and append state annual data
data_list =[]
for state in states:  
    state_data = requests.get(beginning + state+ end).json()
    state_df=pd.DataFrame(state_data["series"][0]["data"],columns = Columns)
    state_df["state_name"] = state
    data_list.append(state_df)
```

```python
#concat the data into a dataframe 
state_full_data=pd.concat(data_list)
```

```python
#send to CSV
state_full_data.to_csv("states_all.csv", index=False, header=True)
```
